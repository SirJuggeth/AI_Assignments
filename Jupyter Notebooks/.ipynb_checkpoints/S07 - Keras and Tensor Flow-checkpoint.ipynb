```python

# Create data frame from raw data set
```

```python
df = pd.read_csv('../DATA/fake_reg.csv')
df.head()
```

```python
sns.pairplot(df)
```

```python

# Create numpy arrays for labels and features from full data set, and then split into data to train model with and data that 
# can be used to test the model with after it has been trained. Typically, labels (what you are wanting model to predict) 
# are for the y axis and features (data model make predictions from/on) are for x.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values
y = df['price'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train.shape
```

```python
type(X_train)
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python

# Choose scale that model will work within and normalize data to it.
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
 help(MinMaxScaler)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

X_train #between zero and one
```

```python

# Creating and training a model
```

```python
pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
help(Sequential)
```

```python
help(Dense)
```

```python
# example of populating model with layers at time of creation
#model = Sequential([Dense(4,activation = 'relu'), 
 #                   Dense(2,activation = 'relu'), 
  #                  Dense(1)])
```

```python
model = Sequential() # Create empty model and add layers individually. More convenient if you want to deactivate a single layer if you want to rebuild the model.

model.add(Dense(4,activation = 'relu')) # number of inputs to a layer is set by what it's given. Number 4 here means that layer will have 4 output neurons to the next layer.
model.add(Dense(4,activation = 'relu'))
model.add(Dense(4,activation = 'relu'))
model.add(Dense(1)) # Last layer, with one output for the estimated price.

model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
loss_df = pd.DataFrame(model.history.history) # Shows progress over training as MSE - how far off the model's predictions were, on average, after each epoch of training. 
loss_df.plot()
```

```python

# Testing the model we created and trained.
```

```python
model.evaluate(X_test, y_test, verbose=0) # Evaluating model's predictions using input data it's never seen before. Returns metric loss of given parameters x&y (in this case Mean Squared Error MSE - set above^ when compiling model. for this model mse should end up around 24~26 after about 250 epochs.).
```

```python
model.evaluate(X_train, y_train, verbose=0) 
```

```python
test_predictions = model.predict(X_test) # Returns numpy array of predictions for given data
```

```python
type(test_predictions)
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,)) #turning numpy array into panda series.
type(test_predictions)
```

```python
test_predictions
```

```python

# Compare predictions to actual values and show how far off the model was
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
pred_df = pd.concat([pred_df,test_predictions],axis=1)
pred_df.columns = ['Test Sets True Y', 'Models Predictions']
pred_df
```

```python
sns.scatterplot(x='Test Sets True Y', y='Models Predictions', data=pred_df)
```

```python

# Showing how far off the model is
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(pred_df['Test Sets True Y'],pred_df['Models Predictions']) # on average, how far off is the model's price predictions
```

```python
df.describe()
```

```python
mean_squared_error(pred_df['Test Sets True Y'],pred_df['Models Predictions']) # same as what was used before when we evaluated the model, because MSE was chosen when the model was compiled.
```
