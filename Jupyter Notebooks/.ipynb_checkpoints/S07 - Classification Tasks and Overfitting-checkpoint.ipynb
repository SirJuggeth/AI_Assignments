```python

# Data prep

X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test) # NOTE: Only training data is to be fitted so as not to impose assumptions.
```

```python
X_train.shape
```

```python

# Model Creation and compilation

model = Sequential()

model.add(Dense(30, activation = 'relu')) # ReLU outputs the node's input value if it is positive, and zero otherwise.
model.add(Dense(15, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid')) # sigmoid is used for output layer because this is a binary classification problem.
# The activation function is what is used to determine if a node will be activated by the neurons connected to 
# its input(sum of all neurons(xw+b)) processed by the function exceeding its threshold value.
# Binary step functions cannot use backpropogation process.

model.compile(loss = 'binary_crossentropy', optimizer='adam')
```

```python

# Train model

model.fit(x=X_train, y=y_train,epochs=600, validation_data=(X_test, y_test))
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
# Shows decrease of model's loss over ran training data epochs, and with validation data.
# At point where validation loss maintains higher values than loss with training data
# than then we know the model has been overfit to the training data.
```

```python

# Redefine (reset) model

model = Sequential()

model.add(Dense(30, activation = 'relu')) # ReLU outputs the node's input value if it is positive, and zero otherwise.
model.add(Dense(15, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid')) # sigmoid is used for output layer because this is a binary classification problem.
# The activation function is what is used to determine if a node will be activated by the neurons connected to 
# its input(sum of all neurons(xw+b)) processed by the function exceeding its threshold value.
# Binary step functions cannot use backpropogation process.

model.compile(loss = 'binary_crossentropy', optimizer='adam')
```

```python
# from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train,epochs=600, validation_data=(X_test, y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
# from tensorflow.keras.layers import Dropout
```

```python

# Using Dropout layers

# Redefine (reset) model

model = Sequential()

model.add(Dense(30, activation = 'relu')) # ReLU outputs the node's input value if it is positive, and zero otherwise.
model.add(Dropout(0.5)) # We set to a rate of 0.5, meaning it will drop 50% of previous layer's 
#                             30 neurons will be turned off randomly, meaning they won't be updated 
#                                 that epoch. (typical values are 0.2-0.5)

model.add(Dense(15, activation = 'relu'))
model.add(Dropout(0.5))

# Binary Classification
model.add(Dense(1, activation = 'sigmoid')) # sigmoid is used for output layer because this is a binary classification problem.
# The activation function is what is used to determine if a node will be activated by the neurons connected to 
# its input(sum of all neurons(xw+b)) processed by the function exceeding its threshold value.
# Binary step functions cannot use backpropogation process.

model.compile(loss = 'binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train,epochs=600, validation_data=(X_test, y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()

# Plot shows significantly improved performance of our model.
```

```python
# model.predict_classes(X_test) <- predict_classes in depricated after Keras 2.5

(model.predict(X_test) > 0.5).astype("int32")
```

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
```

```python
# from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions)) # comparing expected results to model's predictions.
```

```python
confusion_matrix(y_test, predictions) # This shows our network only had one misclassified prediction.
```
