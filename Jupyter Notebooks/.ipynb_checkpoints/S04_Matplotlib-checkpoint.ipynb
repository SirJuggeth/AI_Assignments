```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,100)
y = 2*x
plt.plot(x,y)
plt.title('Straight and Rising Line')
plt.xlabel('Things going sideways')
plt.ylabel('Things going up')
plt.xlim(0, 200)
plt.ylim(0, 250)
plt.savefig('My_first_pyplot.png')
#plt.show()
```

```python
fig = plt.figure() #this creates a blank canvas with the size of 10x10 inches (720x720 pixels)
axes = fig.add_axes([0,0,0.5,1]) # each number is a ratio, 0-1. First 2 num give coords for bottom left corner 
axes.plot(x,y)                                #(ex. 0.5,0.5 would set point of reference at centre of plot)
                                # last 2 nums are how much, on the x and y axis, the plot will take up
                                #(ex. 0.5,0.5 would use half of the canvas space, starting from the indicated bottom left corner.)
```

```python
#DATA
a = np.linspace(0,10,11)
b=a**4
x=np.arange(1,10)
y=x*2
fig = plt.figure(figsize=(9,3), dpi=200) # 9"x3"
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.6,0.25,0.25])
axes1.plot(x,y)
axes1.set_xlim(0,8)
axes1.set_ylim(0,20)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Power of 4')
axes2.plot(a,b)
axes2.plot(x,y)
axes2.set_xlim(0,10)
axes2.set_ylim(0,8000)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed')
plt.show()

fig.savefig('parameterized_figure.jpg', bbox_inches='tight')
```

```python
# NOTE! This returns 2 dimensional array
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8), dpi=200) # **fig_kw are figure keywords, meaning any figure.something arguments can be used here, like figsize and dpi.

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(((x+3)*5),(y**(1/y)))
axes[1][0].plot(b,a)

# Use left,right,top, bottom to stretch subplots
# Use wspace,hspace (width, height) to add spacing between subplots
fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.1,)

# SET YOUR AXES PARAMETERS FIRST

# Parameters at the axes level
axes[0][0].set_title("0 0's Title")

axes[1][1].set_title("1 1's Title")
axes[1][1].set_xlabel("1 1's X Label")

# THEN SET OVERALL FIGURE PARAMETERS

# Parameters at the Figure level
fig.suptitle("I'm up here Level",fontsize=16)

# Save a copy
fig.savefig('subplots.png',bbox_inches='tight')

plt.show()
```
